## joining tables  with common data

```python
#joining prices and securities
pd.read_sql_query("""SELECT psa.symbol, s."GICS Sector", psa.date, s.CIK, s."Address of Headquarters"
                     FROM "prices-split-adjusted" AS psa 
                      INNER JOIN securities AS s ON psa.symbol = s."Ticker symbol" ;""", conn)
```

```python
#adding fundamentals to the table
pd.read_sql_query("""SELECT psa.symbol, s."GICS Sector", psa.date, s.CIK, s."Address of Headquarters",
                    f."Total Assets", f."Total Revenue", psa.open, psa.close
                    FROM "prices-split-adjusted" AS psa
                    INNER JOIN securities AS s ON psa.symbol = s."Ticker symbol"
                    INNER JOIN fundamentals AS f ON psa.symbol = f."Ticker symbol";""", conn)
```

```python
#naming the joined tables new york
pd.read_sql_query("""
    CREATE TABLE "new_york_stock" AS
    SELECT psa.symbol, 
           s."GICS Sector", 
           psa.date, 
           s.CIK, 
           s."Address of Headquarters",
           f."Total Assets", 
           f."Total Revenue", 
           psa.open, 
           psa.close
    FROM "prices-split-adjusted" AS psa
    INNER JOIN securities AS s ON psa.symbol = s."Ticker symbol"
    INNER JOIN fundamentals AS f ON psa.symbol = f."Ticker symbol";
""", conn)
```

```python
#checking if the table has been created
query = "SELECT * FROM new_york_stock LIMIT 20;"
table_contents = pd.read_sql_query(query, conn)
print(table_contents)
```
